```python
import os
import matplotlib.pyplot as plt

# Define the paths to the folders containing the ultrasound images
benign_path = '/kaggle/input/breast-ultrasound-images-dataset/Dataset_BUSI_with_GT/benign'
malignant_path = '/kaggle/input/breast-ultrasound-images-dataset/Dataset_BUSI_with_GT/malignant'
normal_path = '/kaggle/input/breast-ultrasound-images-dataset/Dataset_BUSI_with_GT/normal'

# Initialize counts for each class
benign_count = 0
malignant_count = 0
normal_count = 0

# Loop through each folder and count the number of ultrasound images
for folder_path in [benign_path, malignant_path, normal_path]:
    for file_name in os.listdir(folder_path):
        if 'mask' not in file_name:  # ignore masks
            if 'benign' in folder_path:
                benign_count += 1
            elif 'malignant' in folder_path:
                malignant_count += 1
            elif 'normal' in folder_path:
                normal_count += 1


color1 = (247, 202, 201)  # rose
color2 = (207, 203, 206) # grey
color3 = (219, 254, 217)  # mint green
color1_norm = tuple(c / 255 for c in color1)
color2_norm = tuple(c / 255 for c in color2)
color3_norm = tuple(c / 255 for c in color3)

# Create a pie chart based on the counts
labels = ['Benign', 'Malign', 'Normal']
sizes = [benign_count, malignant_count, normal_count]
colors = [color2_norm, color1_norm, color3_norm]
explode = (0.1, 0, 0)  # optional: add some space between the slices

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt

# Define the paths to the folders containing the ultrasound images
benign_path = '/kaggle/input/breast-ultrasound-images-dataset/Dataset_BUSI_with_GT/benign'
malignant_path = '/kaggle/input/breast-ultrasound-images-dataset/Dataset_BUSI_with_GT/malignant'
normal_path = '/kaggle/input/breast-ultrasound-images-dataset/Dataset_BUSI_with_GT/normal'


# Initialize counts for ultrasounds and masks in each class
benign_ultrasound_count = 0
benign_mask_count = 0
malignant_ultrasound_count = 0
malignant_mask_count = 0
normal_ultrasound_count = 0
normal_mask_count = 0

# Loop through each folder and count the number of ultrasounds and masks
for folder_path in [benign_path, malignant_path, normal_path]:
    for file_name in os.listdir(folder_path):
        if 'mask' in file_name:
            if 'benign' in folder_path:
                benign_mask_count += 1
            elif 'malignant' in folder_path:
                malignant_mask_count += 1
            elif 'normal' in folder_path:
                normal_mask_count += 1
        else:
            if 'benign' in folder_path:
                benign_ultrasound_count += 1
            elif 'malignant' in folder_path:
                malignant_ultrasound_count += 1
            elif 'normal' in folder_path:
                normal_ultrasound_count += 1

# Set the data for the grouped bar chart
ultrasound_counts = [benign_ultrasound_count, malignant_ultrasound_count, normal_ultrasound_count]
mask_counts = [benign_mask_count, malignant_mask_count, normal_mask_count]
ind = np.arange(3)
width = 0.35

# Create the grouped bar chart
p1 = plt.bar(ind, ultrasound_counts, width, color=color1_norm)
p2 = plt.bar(ind + width, mask_counts, width, color=color2_norm)

# Add chart labels and legend
plt.ylabel('Count')
plt.title('Number of Ultrasounds and Masks by Class')
plt.xticks(ind + width / 2, ('Benign', 'Malign', 'Normal'))
plt.legend((p1[0], p2[0]), ('Ultrasounds', 'Masks'))

# Show the chart
plt.show()
```
